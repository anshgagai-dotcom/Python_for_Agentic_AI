Concept 2 — indexing


Definition:
Indexing means accessing (getting) a single item from a list using its position number (index).

Every item in a list has its own position number.

Python starts counting from 0, not 1.

```python
["lipstick", "spray", "bindi", "nailpaint"]
    0            1        2          3
   -4           -3       -2         -1
```

```python
makeup_pouch[2]
```

```python
makeup_pouch[-2]
```

```python
makeup_pouch[0]
```

Concept 3 — IndexError


Definition:
IndexError occurs when you try to access an index (position) that does not exist in a list.

In simple words:
If you ask Python for an item that isn't in the list, Python raises an IndexError.

```python
makeup_pouch[5]
```

Concept 4 — slicing

Definition:
Slicing means taking a part (portion) of a list instead of the whole list.

```python
["lipstick", "spray", "bindi", "nailpaint"]
    0            1        2          3
   -4           -3       -2         -1
```

```python
(makeup_pouch[1:4])
```

```python
(makeup_pouch[0:2])
```

```python
(makeup_pouch[:3])
```

```python
(makeup_pouch[2:])
```

```python
(makeup_pouch[:])
```

```python
(makeup_pouch[-3:])
```

```python
(makeup_pouch[:-2])
```
